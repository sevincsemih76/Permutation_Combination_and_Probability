# Heads or Tail

```python
import random

coin = ('H', 'T')

random.choice(coin)
```

```python
results = {'H':0, 'T':0}

for i in range(100000):
    results[random.choice(list(results.keys()))] += 1
    
print('P(Heads):', results['H']/sum(results.values()))
print('P(Tails):', results['T']/sum(results.values()))
```

```python
# When we increased toss, we got a result close to 50%.
```

# Rolling 2 Dice

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(51)
```

```python
d1 = np.array([1,2,3,4,5,6])
d2 = np.array([1,2,3,4,5,6])
```

```python
dice_1 = []
dice_2 = []

sums = []

for i in range(10000):
    roll1 = np.random.choice(d1)
    roll2 = np.random.choice(d2)
    dice_1.append(roll1)
    dice_2.append(roll2)
    sums.append(roll1 + roll2)
```

```python
print(dice_1)
print(dice_2)
print(sums)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey = True, figsize=(12,4))
sns.countplot(dice_1, ax=ax1)
sns.countplot(dice_2, ax=ax2)
```

```python
# When we increase the shot, an image is formed closer to the normal distribution.
```

```python
sns.countplot(sums)
```

```python
# The probability that the sum of the two dice is 7 is greater than any other total.
# The least likely dice totals are 2 and 12. 
# What is important to note here is that there is a high probability of a total of 7,
# not a certain combination that gives the value 7.
```

```python
for i in set(sums):
    print("P(%d): %.4f" % (i, sums.count(i)/len(sums)))
```

```python
print('P(12): ', sums.count(12)/len(sums))
```

## Permutations

```python
import math

math.factorial(4)
```

```python
def permutation(n,r):
    return math.factorial(n)/math.factorial(n-r)
# we created a function to calculate permutation
```

```python
perm = it.permutations('GRYB', 2)

for i in list(perm):
    print(i)
# Possible outcomes of getting green, red, yellow, and blue candies two at a time
# order counts in
```

```python
permutation(4,2)
```

## Combinations

```python
def combination(n,r):
    return math.factorial(n)/(math.factorial(r) * math.factorial(n-r))
# Possible outcomes of getting green, red, yellow, and blue candies two at a time
# order of selection does not matter
```

```python
combination(5,3)
```

```python
permutation(5,3)
```

```python
comb = it.combinations('GRYB', 2)
```

```python
for i in list(comb):
    print(i)
# match in combination must be unique
```

```python
comb2 = it.combinations('GGYB', 2)
```

```python
for i in list(comb2):
    print(i)
```

```python
comb3 = it.combinations_with_replacement('GGYB', 2)
# combinations; If the input items are unique, there will be no repeat value in every combination.
# But here there are similar ones and the result:
```

```python
for i in list(comb3):
    print(i)
```

```python
a = list(it.product("ABCD", repeat = 2))
print(a)
```

```python
b = list(it.permutations("ABCD", 2))
print(b)
```

```python
c = list(it.combinations("ABCD", 2))
print(c)
# combinations; if the input elements are unique, there will be no repeat values in each combination.
```
